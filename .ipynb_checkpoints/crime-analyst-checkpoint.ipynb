### Data Insight

#### 1. Daerah/lingkungan kota Montreal manakah yang memiliki jumlah kejahatan yang dilaporkan lebih tinggi?

```python
top_daerah = pd.DataFrame(crime['neighbourhood'].value_counts()).rename({"neighbourhood":"Laporan Masuk"}, axis = 1).rename_axis("Daerah/Distrik").head(10)
top_daerah.style.bar()
```

```python
top_daerah = top_daerah.sort_values(by='Laporan Masuk', ascending=False).head(10)

plt.figure(figsize = (10,8))

sns.barplot(data=top_daerah, x=top_daerah.index, y='Laporan Masuk', palette='Reds')

plt.xlabel('Daerah / Distrik')
plt.ylabel('Laporan Masuk')
plt.title('Top 10 Daerah yang paling sering melapor atas tindakan kejahatan')

plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show
```

#### 2. Kejahatan apa yang paling sering terjadi di lingkungan sekitar (kota Montreal)?

```python
#deklarasi 10 data kejahatan
top_crime = crime['neighbourhood'].value_counts().head(10).index

#filter data 
data_top_crime = crime[crime['neighbourhood'].isin(top_crime)]

crime_order = data_top_crime.groupby('neighbourhood')['category'].count().sort_values(ascending=False).index

plt.figure(figsize = (10,6))
sns.countplot(data=data_top_crime, x='neighbourhood', hue='category', order=crime_order, palette='viridis')

plt.xlabel('Distrik')
plt.ylabel('Count')
plt.title('Representasi 10 Distrik yang sering terjadi aksi kejahatan')

plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show()
```

Kejahatan pencurian didalam rumah merupakan category kejahatan yang paling sering terjadi kecuali di distrik Saint-Laurent dan Saint-Leonard.

### 3. Jumlah kejahatan yang dilaporkan dilakukan setiap tahun di Montreal?

```python
pd.DataFrame(crime['year'].value_counts()).rename({"year" : "Laporan Masuk" }, axis=1).rename_axis("tahun")
```

Dapat dilihat bahwa jumlah terjadinya aksi kejahatan dari tahun ke tahun mengalami trend/grafik penurunan yang cukup signifikan, hal ini bisa diliaht pada tahun 2015 dimana terjadi sebanyak 24.444 aksi kejahatan hingga pada tahun 2021 turun menjadi 10.547 aksi kejahatan

### 4.Bagaimana tren umum kejahatan yang dilaporkan pada tahun 2015-2021?

```python
## Plot
plt.figure(figsize = (10, 6))
sns.countplot(data=crime, x='year', hue='category', palette='viridis')

# Labeling
plt.xlabel('Tahun')
plt.ylabel('Count')
plt.title('Distribusi aksi kejahatan dari tahun ke tahun')

plt.tight_layout()
plt.show()
```

Aksi Kejahatan pencurian didalam rumah merupakan aksi kejahatan yang paling sering terjadi di Kota Montreal, kemudian disusul dengan aksi kejahatan begal/theft from motorcycle. Namun, pada tahun 2021 aksi kejahatan yang paling sering dilaporkan oleh masyarakat adalah begal.

```python
crime['Month'] = crime['date'].apply(lambda time : time.month)
crime['Day Of Week'] = crime['date'].apply(lambda time: time.dayofweek)

dmap = {0 : 'Senin', 1 : 'Selasa', 2 : 'Rabu', 3 : 'Kamis', 4 : 'Jumat', 5 : 'Sabtu', 6 : 'Ahad'}
crime['Day Of Week'] = crime['Day Of Week'].map(dmap)

sns.countplot(x='Day Of Week', data=crime, hue='category', palette='viridis')
plt.legend(bbox_to_anchor = (1.05, 1), loc=2, borderaxespad = 0.)
```

Hari senin adalah hari yang paling sering terjadi aksi kejahatan pencurian didalam rumah, hal ini bisa saja terjai karena memang hari seni merupak hari pertam kerja setelah akhir pekan, sehingga potensi maling lebih besar untuk merampok didalam rumah ketika penghuni rumah sedang tidak berada didalam rumah.

```python
#Trend aksi kejahatan dari tahun ketahun

# Hitung total jumlah kejahatan (group by)
crime_years = crime.groupby('year').sum(numeric_only=True).drop(['longitude', 'latitude'], axis=1).rename({'count': 'Laporan Masuk'}, axis=1)


plt.figure(figsize=(10, 6))
sns.lineplot(data=crime_years, x=crime_years.index, y='Laporan Masuk')

# Labeling
plt.xlabel('Year')
plt.ylabel('Laporan Masuk')
plt.title('Trend aksi kejahatan 2015 - 2021')

plt.tight_layout()
plt.show()
```

Bulan July dan bulan Agustus aksi kejahatan begal sering terjadi. 

```python
# Korelasi antara hari dan bulan

xy = crime.groupby(by=['Day Of Week', 'Month']).count()['category'].unstack()
xy.head()
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(xy, cmap='coolwarm')
```
